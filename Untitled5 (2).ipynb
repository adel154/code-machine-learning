```python
# Charger le fichier "Shoot1.csv"
filename_shoot = 'Shoot1.csv'
headers = ['aX', 'aY', 'aZ']
df_shoot = pd.read_csv(filename_shoot, names=headers, skipinitialspace=True, sep=',', keep_default_na=False)
df_shoot['aX'] = pd.to_numeric(df_shoot['aX'], errors='coerce')
df_shoot['aY'] = pd.to_numeric(df_shoot['aY'], errors='coerce')
df_shoot['aZ'] = pd.to_numeric(df_shoot['aZ'], errors='coerce')

# Charger le fichier "no_shout11.csv"
filename_no_shout = 'no_shout11.csv'
df_no_shout = pd.read_csv(filename_no_shout, names=headers, skipinitialspace=True, sep=',', keep_default_na=False)
df_no_shout['aX'] = pd.to_numeric(df_no_shout['aX'], errors='coerce')
df_no_shout['aY'] = pd.to_numeric(df_no_shout['aY'], errors='coerce')
df_no_shout['aZ'] = pd.to_numeric(df_no_shout['aZ'], errors='coerce')
```

```python
# Sélectionner les colonnes pertinentes
X_shoot = df_shoot[['aX', 'aY', 'aZ']]
X_no_shout = df_no_shout[['aX', 'aY', 'aZ']]

# Standardiser les données
scaler_shoot = StandardScaler()
X_shoot_scaled = scaler_shoot.fit_transform(X_shoot)

scaler_no_shout = StandardScaler()
X_no_shout_scaled = scaler_no_shout.fit_transform(X_no_shout)

```

```python
# Supprimez les lignes contenant des NaN
X_shoot_scaled = X_shoot_scaled[~np.isnan(X_shoot_scaled).any(axis=1)]
X_no_shout_scaled = X_no_shout_scaled[~np.isnan(X_no_shout_scaled).any(axis=1)]

# Vérifiez à nouveau s'il reste des NaN
print(np.isnan(X_shoot_scaled).sum())
print(np.isnan(X_no_shout_scaled).sum())
```

```python
# Définir l'auto-encodeur pour Shoot1.csv
input_dim = X_shoot_scaled.shape[1]
encoding_dim = 2

input_layer = keras.layers.Input(shape=(input_dim,))
encoded = keras.layers.Dense(encoding_dim, activation='tanh')(input_layer)
decoded = keras.layers.Dense(input_dim, activation='sigmoid')(encoded)

autoencoder_shoot = keras.models.Model(input_layer, decoded)
encoder_shoot = keras.models.Model(input_layer, encoded)

# Compiler le modèle avec un taux d'apprentissage adapté
autoencoder_shoot.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss='mse')

# Ajuster les paramètres d'entraînement
autoencoder_shoot.fit(X_shoot_scaled, X_shoot_scaled, epochs=100, batch_size=32, shuffle=True, validation_split=0.25)



```

```python
# Visualiser les erreurs
predictions = autoencoder_shoot.predict(X_shoot_scaled)
errors = X_shoot_scaled - predictions

plt.hist(errors.flatten(), bins=50)
plt.xlabel('Error')
plt.ylabel('Count')
plt.title('Histogram of Prediction Errors')
plt.show()
```

```python
# Fonction pour extraire les poids et les biais et les sauvegarder dans un fichier
def weights_to_cpp(model, filename="weights_and_biases.txt"):
    model.summary()  # Affiche le résumé du modèle pour vérification

    weights = []
    biases = []

    # Extraire les poids et les biais de chaque couche du modèle
    for layer in model.layers:
        if len(layer.get_weights()) > 0:  # Vérifier si la couche a des poids
            W, B = layer.get_weights()
            weights.append(W.flatten())
            biases.append(B.flatten())

    # Convertir les poids et les biais en une liste simple de valeurs
    z = [w for weight in weights for w in weight]
    b = [bias for bias_array in biases for bias in bias_array]

    # Écrire les poids, les biais et l'architecture dans un fichier
    with open(filename, "w") as f:
        f.write("weights: {")
        for i in range(len(z)):
            if i < len(z)-1:
                f.write(str(z[i]) + ", ")
            else:
                f.write(str(z[i]))
        f.write("}\n\n")

        f.write("biases: {")
        for i in range(len(b)):
            if i < len(b)-1:
                f.write(str(b[i]) + ", ")
            else:
                f.write(str(b[i]))
        f.write("}\n\n")

        # Extraire l'architecture du modèle
        arch = []
        if hasattr(model.layers[0], 'input_shape') and len(model.layers[0].input_shape) > 1:
            arch.append(model.layers[0].input_shape[1])
        for layer in model.layers:
            if hasattr(layer, 'output_shape') and len(layer.output_shape) > 1:
                arch.append(layer.output_shape[1])

        f.write("Architecture: {")
        for i in range(len(arch)):
            if i < len(arch)-1:
                f.write(str(arch[i]) + ", ")
            else:
                f.write(str(arch[i]))
        f.write("}\n")

    print("Architecture (alpha):", arch)
    print("Layers:", len(arch))
    print("Weights:", z)
    print("Biases:", b)

# Appel de la fonction pour extraire les poids et les biais pour les deux modèles
weights_to_cpp(autoencoder_shoot, "weights_and_biases_shoot.txt")
weights_to_cpp(autoencoder_no_shout, "weights_and_biases_no_shout.txt")




```
